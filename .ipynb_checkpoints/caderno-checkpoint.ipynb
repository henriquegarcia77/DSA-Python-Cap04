```python
conjunto_a = {1, 2, 3, 4}
conjunto_b = {3, 4, 5, 6}
```

```python
uniao = conjunto_a.union(conjunto_b)
print(uniao)
```

```python
intersecao = conjunto_a.intersection(conjunto_b)
print(intersecao)
```
