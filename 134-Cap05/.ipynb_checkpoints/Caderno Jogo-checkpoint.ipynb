```python
#estilizar entrada do jogo
print("-"*25)
print("--Seja Bem-Vindo--")
print("-"*25)

#instrução para comando
print("Para participar digite: 'pedra', 'papel' ou 'tesoura'")
print("-"*25)

#salvar as possibilidade de jogo
jogadas = ("pedra","papel","tesoura")

#receber comando
jogador1_entrada = input("Jogador 1 digite sua jogada:")
jogador2_entada = input("Jogador 2 digite sua jogada:")

#tratar o que foi recebido
jogador1 = jogador1_entrada.lower().strip()
jogador2 = jogador2_entada.lower().strip()

#informar a jogada
print(f"Jogador 1 sua jogada foi: {jogador1}")
print(f"Jogador 2 sua jogada foi: {jogador2}")

#ver se o usuario digitou o pedra,papel e tesoura

if jogador1 not in jogadas or jogador2 not in jogadas:
    print("Digite um valor válido (pedra, papel ou tesoura)")

#verificar se teve empate
elif jogador1 == jogador2:
    print("Empate!")

#se o jogador 1 jogar um dessas possibilidades ele ganha caso vai para o else o jogador 2 ganha
elif jogador1 == "pedra" and jogador2 == "tesoura" or \
     jogador1 == "papel" and jogador2 == "pedra" or \
     jogador1 == "tesoura" and jogador2 == "papel":
     print("Jogador 1 venceu!!")
else:
    print("Jogador 2 venceu!!")

```
